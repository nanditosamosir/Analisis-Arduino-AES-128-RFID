```python
import matplotlib.pyplot as plt

# Membuat gambar dan sumbu
plt.figure(figsize=(10, 6))

# Plot untuk Encryption Time
plt.plot(data['byte_size'], data['encryption_time'], marker='o', label='Encryption Time')
plt.plot(data['byte_size'], data['decryption_time'], marker='o', label='Decryption Time')

# Menambahkan judul dan label sumbu
plt.title('Encryption and Decryption Time vs Byte')
plt.xlabel('Data Size (Bytes)')
plt.ylabel('Time (s)')

# Menambahkan grid dan legenda
plt.grid(True)
plt.legend()

# Menampilkan plot
plt.show()
```

```python
# Menambahkan kolom rasio decryption_time / encryption_time
data['ratio'] = data['decryption_time'] / data['encryption_time']

# Membuat grafik rasio
plt.figure(figsize=(10, 6))

# Grafik 2: Ratio Decryption Time / Encryption Time
plt.plot(data['byte_size'], data['ratio'], marker='o', color='purple', label='Decryption/Encryption Ratio')

# Menambahkan judul, label sumbu, grid, dan legenda
plt.title('Decryption to Encryption Time Ratio vs Byte')
plt.xlabel('Data Size (Bytes)')
plt.ylabel('Ratio (Decryption/Encryption)')
plt.grid(True)
plt.legend()
plt.show()
```
